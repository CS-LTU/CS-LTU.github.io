## Section 1 — The Winner Popup Overlay

When a player goes bankrupt, we create a semi-transparent overlay that covers the entire screen, with a card in the centre showing the winner's name and final fortune.

```css
.winner-popup {
    position: fixed;     /* stays in place even if the page scrolls */
    inset: 0;            /* shorthand for top:0, right:0, bottom:0, left:0 */
    background: rgba(0, 0, 0, 0.85);
    display: flex;
    align-items: center;
    justify-content: center;
    z-index: 999;        /* appears above everything */
    backdrop-filter: blur(8px);  /* blurs the content behind */
}

.winner-card {
    background: var(--surface);
    border: 2px solid var(--gold);
    border-radius: 16px;
    padding: 40px;
    text-align: center;
}
```

## Section 2 — The showWinner Function

This function creates the overlay dynamically and appends it to the page body:

```javascript
function showWinner(w) {
    const div = document.createElement("div");
    div.className = "winner-popup";
    div.innerHTML =
        '<div class="winner-card">' +
        '<h2>🏆 ' + w.name + ' Wins!</h2>' +
        '<p>Final fortune: £' + w.cash +
        ' with ' + w.props.length + ' properties</p>' +
        '<button onclick="location.reload()" class="primary"' +
        ' style="width:auto;padding:10px 30px;">Play Again</button>' +
        '</div>';
    document.body.appendChild(div);
}
```

The "Play Again" button uses `location.reload()`, which simply refreshes the page. Since all game state is held in JavaScript variables (not saved anywhere), refreshing resets everything to the starting position.

## Section 3 — Responsive Design

On smaller screens (below 900px wide), the board and sidebar need to stack vertically instead of sitting side by side. We also shrink the board cells to fit. This is done with a CSS `@media` query:

```css
@media (max-width: 900px) {
    .game-layout {
        flex-direction: column;
        align-items: center;
    }
    .board {
        grid-template-columns: 60px repeat(9, 42px) 60px;
        grid-template-rows: 60px repeat(9, 42px) 60px;
    }
    .cell .name { font-size: 0.35rem; }
    .board-center .title { font-size: 1.2rem; }
}
```

The `@media` rule applies its styles only when the browser window is 900px wide or narrower. Everything inside the rule overrides the normal styles for smaller screens.

## Section 4 — Understanding position: fixed vs position: absolute

Both `fixed` and `absolute` remove an element from the normal document flow, but they behave differently:

| Property | `position: absolute` | `position: fixed` |
|---|---|---|
| Positioned relative to | Nearest positioned ancestor | The browser viewport |
| Scrolls with page | Yes | No (stays in place) |
| Best for | Elements inside a container (colour bars) | Full-screen overlays |

The winner popup uses `fixed` so it always covers the entire screen, even if the page has been scrolled.

## Section 5 — The inset Shorthand

Instead of writing four separate properties:

```css
top: 0;
right: 0;
bottom: 0;
left: 0;
```

You can use the `inset` shorthand:

```css
inset: 0;
```

This is a modern CSS feature that makes the element stretch to fill its positioning context (in this case, the viewport). It is supported in all modern browsers.
