## Section 2 — Architecture Review

The complete game has four main sections, all in a single HTML file:

**1. CSS (inside `<style>`)**
- CSS variables for the colour palette (`:root`)
- Board grid layout (`.board`, `.cell`, `.color-bar`)
- Sidebar panels (`.panel`, `.stat`, `.turn-badge`)
- Token and ownership indicator styles
- Button styles with gradients and transitions
- Game log with colour-coded lines
- Winner popup overlay
- Responsive media query

**2. HTML (inside `<body>`)**
- Header with title and subtitle
- Game layout container (flexbox: board + sidebar)
- Board wrapper with an empty `<div>` for JavaScript to populate
- Sidebar with player panels, action buttons, and game log

**3. Board Data (inside `<script>`)**
- `B` array: 40 board spaces with compact property names
- `CHANCE` array: 7 chance cards with action functions
- `CHEST` array: 7 community chest cards with action functions

**4. Game Logic (inside `<script>`)**
- `buildBoard()`: creates all 40 cells and the centre area
- `boardToGrid(i)` and `getSide(i)`: mapping helpers
- `G` object: the game state (players, current turn, phase, game-over)
- `rollDice()`: dice, jail, movement, landing
- `handleLanding()`: switch on space type
- `buyProperty()` and `endTurn()`: player actions
- `owner()` and `checkBust()`: utility functions
- `updateAll()`: refreshes the entire display
- `log()`: adds messages to the game log
- `showWinner()`: creates the victory popup

## Section 3 — The Data Flow

Every game action follows the same pattern:

```
User clicks button
    → Game function modifies the state (G object)
        → updateAll() reads the state and refreshes the display
```

For example, when the user clicks "Roll Dice":

1. `rollDice()` is called
2. It generates random dice values
3. It updates the player's position in `G.players[G.cur].pos`
4. It calls `handleLanding()` which may change cash, properties, or jail status
5. It sets `G.phase` to indicate what the player can do next
6. It calls `updateAll()`
7. `updateAll()` redraws tokens, updates stats, shows ownership dots, and enables the correct buttons

This "modify state then refresh display" pattern is the foundation of almost every interactive web application.
