## Section 1 — The 11x11 Grid

The board has 11 columns and 11 rows. The corner cells sit in the first and last row/column and are wider (72px). The 9 cells along each edge are narrower (56px). Here is the CSS:

```css
.board {
    display: grid;
    grid-template-columns: 72px repeat(9, 56px) 72px;
    grid-template-rows:    72px repeat(9, 56px) 72px;
    gap: 1px;
}
```

The `repeat(9, 56px)` creates 9 columns of 56px each. The first and last columns are 72px for the corners. The `gap: 1px` adds thin lines between cells, like the borders on a real board.

## Section 3 — How boardToGrid Works

The `boardToGrid` function is the key to the entire visual board. Let us trace through it step by step:

```javascript
function boardToGrid(i) {
    if (i <= 10) return { row: 10, col: 10 - i };
    if (i <= 20) return { row: 10 - (i - 10), col: 0 };
    if (i <= 30) return { row: 0, col: i - 20 };
    return { row: i - 30, col: 10 };
}
```

**Bottom row (spaces 0 to 10):** All on row 10. GO is at column 10 (far right), and the column decreases as `i` increases, so the spaces run right to left.

**Left column (spaces 11 to 20):** All on column 0. Pall Mall (space 11) is at row 9 (near the bottom), and the row decreases as `i` increases, so the spaces run bottom to top.

**Top row (spaces 21 to 30):** All on row 0. Strand (space 21) is at column 1, and the column increases as `i` increases, so the spaces run left to right.

**Right column (spaces 31 to 39):** All on column 10. Regent Street (space 31) is at row 1 (near the top), and the row increases as `i` increases, so the spaces run top to bottom.

## Section 4 — CSS Grid: 1-Based vs 0-Based Indexing

Notice this line in the code:

```javascript
cell.style.gridRow = g.row + 1;
```

CSS Grid uses **1-based** indexing (the first row is row 1, not row 0). JavaScript arrays and our `boardToGrid` function use **0-based** indexing. So we add 1 when setting the grid position. This is a common source of bugs, so keep it in mind.

## Section 5 — The Centre Area

The board centre spans columns 2 through 10 and rows 2 through 10. In CSS Grid, this is written as:

```css
.board-center {
    grid-column: 2 / 11;   /* start at column 2, end before column 11 */
    grid-row: 2 / 11;       /* start at row 2, end before row 11 */
}
```

The end value is **exclusive**, so `2 / 11` means columns 2, 3, 4, 5, 6, 7, 8, 9, and 10 (nine columns). This creates a large area in the middle of the board where we will put the game title and dice.

## Section 6 — Identifying Which Side a Cell Is On

Each cell needs to know which side of the board it sits on, because the colour bar (the property group strip) goes on a different edge depending on the side. We determine this with a simple helper:

```javascript
function getSide(i) {
    if (i <= 10) return "side-bottom";
    if (i <= 20) return "side-left";
    if (i <= 30) return "side-top";
    return "side-right";
}
```

This will become important in the next workshop when we add colour bars.

````python
# Let us verify the mapping with a quick Python check
def board_to_grid(i):
    if i <= 10: return (10, 10 - i)
    if i <= 20: return (10 - (i - 10), 0)
    if i <= 30: return (0, i - 20)
    return (i - 30, 10)

# Check the four corners
corners = {0: "GO", 10: "Visiting", 20: "Free Parking", 30: "Go To Jail"}
for idx, name in corners.items():
    row, col = board_to_grid(idx)
    print(f"Space {idx:2d} ({name:12s}) -> row={row}, col={col}")

print()

# Check a few edge cells
edges = {1: "Old Kent Rd", 15: "Marylebone", 25: "Fenchurch", 37: "Park Lane"}
for idx, name in edges.items():
    row, col = board_to_grid(idx)
    print(f"Space {idx:2d} ({name:12s}) -> row={row}, col={col}")
````
