## Section 2 — Understanding the Board Layout

A Monopoly board is a square with properties around the edges and an empty centre. We can represent this as an **11 by 11 CSS Grid**:

- The **4 corners** are larger cells (GO, Visiting, Free Parking, Go To Jail)
- The **4 edges** each have 9 smaller cells for properties and special squares
- The **centre** (a 9x9 area) holds the game title and dice display

The 40 board positions map to grid coordinates like this:
- **Bottom row** (spaces 0 to 10): row 11, columns 11 down to 1 (right to left)
- **Left column** (spaces 11 to 20): column 1, rows 10 down to 1 (bottom to top)
- **Top row** (spaces 21 to 30): row 1, columns 2 to 11 (left to right)
- **Right column** (spaces 31 to 39): column 11, rows 2 to 10 (top to bottom)

## Section 5 — The Mapping Function

The most important new concept in Phase 2 is the **mapping function** that converts a board index (0 to 39) into a grid position (row and column). Here is the logic:

```
Space 0  (GO)          → row 10, col 10  (bottom-right corner)
Space 1  (Old Kent Rd) → row 10, col 9   (bottom row, moving left)
...
Space 10 (Visiting)    → row 10, col 0   (bottom-left corner)
Space 11 (Pall Mall)   → row 9,  col 0   (left column, moving up)
...
Space 20 (Free Parking) → row 0, col 0   (top-left corner)
Space 21 (Strand)       → row 0, col 1   (top row, moving right)
...
Space 30 (Go To Jail)   → row 0, col 10  (top-right corner)
Space 31 (Regent St)    → row 1, col 10  (right column, moving down)
...
Space 39 (Mayfair)      → row 9, col 10  (right column, near bottom)
```

We will implement this as a `boardToGrid(index)` function in JavaScript.
