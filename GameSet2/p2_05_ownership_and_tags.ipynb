## Section 2 — Property Tags in the Sidebar

In the sidebar, each player's properties are shown as small colour-coded tags. The colour matches the property group colour from the board data. For example, if Player 1 owns Old Kent Road, they get a small brown tag reading "Old Kent Rd".

```javascript
const pp = document.getElementById("s_p" + (i+1) + "props");
pp.innerHTML = pl.props.map(function(idx) {
    const s = B[idx];
    return '<span class="prop-tag" style="background:' + (s.col || "#555") + '">' + s.n + '</span>';
}).join("");
```

The `.map()` method creates a new array of HTML strings, one per property. The `.join("")` combines them into a single string. The `s.col || "#555"` provides a default grey colour if the property has no colour defined.

## Section 3 — The CSS for Property Tags

```css
.props-list {
    margin-top: 6px;
    display: flex;
    flex-wrap: wrap;      /* tags wrap to the next line */
    gap: 3px;
}

.prop-tag {
    font-size: 0.6rem;
    padding: 2px 6px;
    border-radius: 4px;
    color: #fff;
    font-weight: 600;
}
```

The `flex-wrap: wrap` is important. As players accumulate properties, the tags need to flow onto multiple lines rather than overflowing the sidebar.

## Section 5 — The querySelector Method

You have seen `document.getElementById()` for finding elements by their ID. The `querySelector` method is more flexible. It uses CSS selector syntax:

| Method | Example | Finds |
|---|---|---|
| `getElementById("x")` | `getElementById("cell-5")` | The element with id="cell-5" |
| `querySelector(".x")` | `cell.querySelector(".owned-indicator")` | The first child with class "owned-indicator" |
| `querySelectorAll(".x")` | `cell.querySelectorAll(".token")` | All children with class "token" |

We use `querySelector` to find existing ownership dots so we can remove them before adding new ones. This prevents dots from piling up.
