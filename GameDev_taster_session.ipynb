## Activity 1 – Simple 1D runner (example)

The player moves left and right along a line to reach the goal.

```python
length = 10         # positions 0..9
player_pos = 0
goal_pos = 9

def show_world():
    world = ["." for _ in range(length)]
    world[player_pos] = "P"
    world[goal_pos] = "G"
    print("".join(world))

print("Reach the goal (G). Use 'l' for left, 'r' for right.")
while True:
    show_world()
    move = input("Move (l/r): ").strip().lower()

    if move == "l" and player_pos > 0:
        player_pos -= 1
    elif move == "r" and player_pos < length - 1:
        player_pos += 1
    else:
        print("Invalid move or edge of world.")

    if player_pos == goal_pos:
        show_world()
        print("You reached the goal! You win!")
        break
```
