## Section 1 — Google Fonts

The final game uses two custom fonts loaded from Google Fonts. Adding a font is as simple as putting an `@import` line inside your `<style>` tag. Professional typography makes a huge difference to how polished your project looks.

```css
@import url('https://fonts.googleapis.com/css2?family=JetBrains+Mono:wght@400;700&family=Space+Grotesk:wght@400;600;700&display=swap');
```

- **JetBrains Mono** is a monospaced font (every character is the same width), perfect for the game console.
- **Space Grotesk** is a clean, modern font used for headings.

## Section 3 — Anatomy of the Complete Game

Let us review how every part connects. The game has three main sections inside a single HTML file:

**1. HTML structure** (the `<body>` tag)
- A title and subtitle
- A two-column grid layout: console on the left, sidebar on the right
- The sidebar contains: turn indicator, dice display, two player cards, and action buttons

**2. CSS styling** (the `<style>` tag)
- CSS variables for the colour palette
- Grid layout for the main container
- Card styles for player information
- Button styles with hover effects and disabled states
- Colour-coded log line classes

**3. JavaScript logic** (the `<script>` tag)
- `BOARD` array: 40 spaces with their types and costs
- `CHANCE_CARDS` and `CHEST_CARDS`: arrays of card objects with action functions
- `state` object: players, current turn, phase, game-over flag
- `rollDice()`: generates dice, handles jail, moves player, calls `handleSpace`
- `handleSpace()`: switch statement for each space type
- `buyProperty()`: deducts cost, adds to player's properties
- `endTurn()`: switches to the next player
- `getPropertyOwner()`: searches for who owns a position
- `checkBankrupt()`: ends the game if cash drops to zero or below
- `log()`: adds a coloured line to the console
- `updateUI()`: refreshes all display elements from the state

## Section 5 — Extension Ideas

Here are some features you could add to make the game your own:

**Easier extensions:**
- Add a "New Game" button that resets everything
- Show a running turn counter in the sidebar
- Add more Chance and Community Chest cards
- Change the colour scheme or try a light theme
- Add sound effects using the `Audio` object

**Medium extensions:**
- Double rent when a player owns all properties of one colour
- Add a "mortgage" feature to raise emergency cash
- Let players choose their names at the start
- Save game state to `localStorage` so it persists across page refreshes

**Harder extensions:**
- Add a visual board display around the edge of the screen
- Support 3 or 4 players
- Add houses and hotels (rent multipliers)
- Create an AI opponent that makes decisions automatically
- Build it as a networked multiplayer game using WebSockets
