## Section 1 — What is HTML?

HTML stands for HyperText Markup Language. It is not a programming language — it is a *markup* language. That means you use special **tags** to describe the structure of content on a page. Your browser reads these tags and renders them as the web page you see.

Every HTML tag looks like this:

```
<tagname>content goes here</tagname>
```

The first tag **opens** and the second (with the `/`) **closes**. Everything between them is the content.

## Section 2 — Set Up Your Project Folder

Before writing any code, create a folder on your computer to keep your workshop files organised. The cell below uses Python to create this for you. If you prefer, you can do it manually in your file explorer instead.

````python
import os

# Create a project folder on your Desktop (adjust the path if needed)
project_folder = os.path.join(os.path.expanduser("~"), "text_monopoly")
os.makedirs(project_folder, exist_ok=True)

print(f"Project folder ready at: {project_folder}")
````

## Section 3 — Write a Minimal HTML Page

Every HTML document has the same skeleton. Let us create one now. The cell below writes a file called `00_first_page.html` into your project folder. After running it, navigate to that folder and double-click the file to open it in your browser.

````python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My First Page</title>
</head>
<body>
    <h1>Hello, World!</h1>
    <p>This is my very first web page.</p>
</body>
</html>
"""

filepath = os.path.join(project_folder, "00_first_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"File saved to: {filepath}")
print("Open this file in your browser to see the result!")
````

## Section 4 — Understanding the Skeleton

Let us break down what each part does:

| Tag | Purpose |
|---|---|
| `<!DOCTYPE html>` | Tells the browser this is an HTML5 document |
| `<html lang="en">` | The root element; `lang="en"` means the content is in English |
| `<head>` | Contains metadata (not visible on the page itself) |
| `<meta charset="UTF-8">` | Allows special characters like £ and emoji |
| `<title>` | The text shown in the browser tab |
| `<body>` | Everything the user can actually see on the page |
| `<h1>` | A top-level heading |
| `<p>` | A paragraph of text |

Every HTML file you create in this workshop series will follow this same skeleton.

## Section 5 — Add More Content

Now let us add headings of different sizes, a list, and a hyperlink. Run the cell below and refresh your browser to see the changes.

````python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My First Page</title>
</head>
<body>
    <h1>Text Monopoly</h1>
    <h2>About the Game</h2>
    <p>Text Monopoly is a simplified version of the classic board game.
       We will build it step by step across eight workshops.</p>

    <h2>What We Will Build</h2>
    <ul>
        <li>A game board with 40 spaces</li>
        <li>Dice rolling</li>
        <li>Buying properties</li>
        <li>Paying rent</li>
        <li>A jail system</li>
    </ul>

    <h3>Useful Links</h3>
    <p>Learn more about HTML at
       <a href="https://developer.mozilla.org/en-US/docs/Web/HTML">MDN Web Docs</a>.
    </p>
</body>
</html>
"""

filepath = os.path.join(project_folder, "00_first_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("File updated! Refresh your browser to see the new content.")
````

## Section 6 — Key Tags You Just Used

Here is a quick reference for the new tags:

- `<h1>` to `<h6>` — headings, from largest to smallest
- `<p>` — a paragraph
- `<ul>` — an unordered (bulleted) list
- `<li>` — a list item (goes inside `<ul>`)
- `<a href="...">` — a hyperlink; the `href` attribute holds the URL

You will use all of these throughout the workshop series.

## Section 7 — Add a Simple Table

Tables are useful for displaying structured data. In our game, we will later use them (or something similar) for the board. For now, let us practise the syntax.

````python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>My First Page</title>
</head>
<body>
    <h1>Text Monopoly</h1>

    <h2>Sample Properties</h2>
    <table border="1">
        <thead>
            <tr>
                <th>Property</th>
                <th>Cost</th>
                <th>Rent</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Old Kent Road</td>
                <td>&pound;60</td>
                <td>&pound;10</td>
            </tr>
            <tr>
                <td>Mayfair</td>
                <td>&pound;400</td>
                <td>&pound;100</td>
            </tr>
        </tbody>
    </table>

    <p><code>&amp;pound;</code> is an HTML entity that displays the &pound; symbol.</p>
</body>
</html>
"""

filepath = os.path.join(project_folder, "00_first_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("File updated with a table! Refresh your browser.")
````
