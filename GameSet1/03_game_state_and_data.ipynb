## Section 2 — The Game State Object

Now we need to track the game itself. Who is playing? Whose turn is it? What phase of the turn are we in? We store all of this in a single `state` object.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Game State Object</title>
    <style>
        :root { --bg: #0d1117; --surface: #161b22; --border: #30363d; --text: #e6edf3; --green: #3fb950; }
        * { margin: 0; padding: 0; box-sizing: border-box; }
        body { background: var(--bg); color: var(--text); font-family: Georgia, serif; padding: 24px; }
        h1 { color: var(--green); margin-bottom: 16px; }
        .card { background: var(--surface); border: 1px solid var(--border); border-radius: 12px; padding: 16px; margin-bottom: 12px; }
        .card h2 { margin-bottom: 8px; font-size: 1.1rem; }
        .stat { display: flex; justify-content: space-between; padding: 4px 0; font-size: 0.9rem; }
        .label { color: #8b949e; }
    </style>
</head>
<body>
    <h1>Game State</h1>
    <div id="cards"></div>

    <script>
        // The game state object - the single source of truth
        let state = {
            players: [
                { name: "Player 1", cash: 1500, position: 0, properties: [], inJail: false, jailTurns: 0 },
                { name: "Player 2", cash: 1500, position: 0, properties: [], inJail: false, jailTurns: 0 }
            ],
            currentPlayer: 0,
            dice: [0, 0],
            phase: "roll",      // "roll", "buy", or "endturn"
            gameOver: false,
            turnCount: 0
        };

        // Build the display using a loop
        const container = document.getElementById("cards");
        for (let i = 0; i < state.players.length; i++) {
            const p = state.players[i];
            const card = document.createElement("div");
            card.className = "card";
            card.innerHTML =
                "<h2>" + p.name + "</h2>" +
                "<div class='stat'><span class='label'>Cash</span><span>£" + p.cash + "</span></div>" +
                "<div class='stat'><span class='label'>Position</span><span>" + p.position + "</span></div>" +
                "<div class='stat'><span class='label'>Properties</span><span>" + p.properties.length + "</span></div>" +
                "<div class='stat'><span class='label'>In Jail</span><span>" + p.inJail + "</span></div>";
            container.appendChild(card);
        }
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "03_game_state.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Game state display saved! Refresh your browser.")
```

## Section 3 — The updateUI Function

In our game, data changes all the time (cash goes up, positions change, properties are bought). We need a single function that reads the current state and updates everything on screen. This pattern of "change data, then refresh the display" is fundamental to game development.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Update UI Pattern</title>
    <style>
        :root { --bg: #0d1117; --surface: #161b22; --border: #30363d; --text: #e6edf3; --green: #3fb950; --blue: #58a6ff; }
        * { margin: 0; padding: 0; box-sizing: border-box; }
        body { background: var(--bg); color: var(--text); font-family: Georgia, serif; padding: 24px; }
        h1 { color: var(--green); margin-bottom: 16px; }
        .cards { display: flex; gap: 16px; margin-bottom: 16px; }
        .card { background: var(--surface); border: 1px solid var(--border); border-radius: 12px; padding: 16px; flex: 1; }
        .stat { display: flex; justify-content: space-between; padding: 4px 0; font-size: 0.9rem; }
        .label { color: #8b949e; }
        button { padding: 10px 20px; border: none; border-radius: 8px; background: var(--green); color: #fff; font-size: 1rem; cursor: pointer; margin-right: 8px; }
    </style>
</head>
<body>
    <h1>Update UI Pattern</h1>
    <div class="cards">
        <div class="card" id="card0"></div>
        <div class="card" id="card1"></div>
    </div>
    <button onclick="giveP1Money()">Give P1 £200</button>
    <button onclick="giveP2Money()">Give P2 £200</button>

    <script>
        let state = {
            players: [
                { name: "Player 1", cash: 1500, position: 0, properties: [] },
                { name: "Player 2", cash: 1500, position: 0, properties: [] }
            ]
        };

        // This function refreshes the entire display from the state
        function updateUI() {
            for (let i = 0; i < state.players.length; i++) {
                const p = state.players[i];
                document.getElementById("card" + i).innerHTML =
                    "<h2>" + p.name + "</h2>" +
                    "<div class='stat'><span class='label'>Cash</span><span>£" + p.cash + "</span></div>" +
                    "<div class='stat'><span class='label'>Position</span><span>" + p.position + "</span></div>";
            }
        }

        // Step 1: change the data. Step 2: call updateUI().
        function giveP1Money() {
            state.players[0].cash += 200;  // change data
            updateUI();                     // refresh display
        }

        function giveP2Money() {
            state.players[1].cash += 200;
            updateUI();
        }

        // Initial render
        updateUI();
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "03_game_state.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Update UI pattern saved! Click the buttons to see cash change.")
```

Card decks add randomness to the game. Each card has a `text` description and an `action` function. The action function receives a player object and modifies it directly. This is a great example of passing objects to functions.

```python
# This cell just shows the data structure - no file write needed.
# We will add it to the full game in later workshops.

print("""Chance and Community Chest cards are arrays of objects.
Each object has:
  - text: what the card says
  - action: a function that modifies the player

Example:
  { text: "Advance to GO! Collect £200",
    action: function(player) { player.position = 0; player.cash += 200; } }

  { text: "Speeding fine: pay £50",
    action: function(player) { player.cash -= 50; } }

JavaScript lets you store functions inside objects.
The (player) => { ... } syntax is called an arrow function.
It does the same thing as function(player) { ... } but shorter.""")
```
