# Workshop 01 — Styling with CSS

Your HTML page works, but it looks plain. In this workshop you will learn **CSS** (Cascading Style Sheets), the language that controls colours, fonts, spacing, and layout. By the end you will have styled the page to look modern and polished, ready to become a game interface.

**What you will learn:**
- How to add CSS to an HTML page
- Selectors, properties, and values
- Colours, fonts, spacing, and borders
- CSS custom properties (variables) for a consistent theme

**Time:** roughly 15 to 20 minutes.

## Section 1 — Three Ways to Add CSS

There are three ways to include CSS in your HTML:

1. **Inline** — directly on an element using the `style` attribute. Quick but messy.
2. **Internal** — inside a `<style>` tag in the `<head>`. Good for single-file projects.
3. **External** — in a separate `.css` file linked with `<link>`. Best for larger projects.

We will use the **internal** method (option 2) throughout this workshop series because it keeps everything in a single file, which is simpler when you are learning.

## Section 2 — Your First Styled Page

Let us take the game page from Workshop 00 and add a dark theme. Notice how CSS rules sit inside the `<style>` tag. Each rule has a **selector** (what to style) and **declarations** (how to style it).

```python
import os

project_folder = os.path.join(os.path.expanduser("~"), "text_monopoly")
os.makedirs(project_folder, exist_ok=True)

html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Text Monopoly - Styled</title>
    <style>
        /* ----- RESET & BASE ----- */
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;   /* makes width calculations simpler */
        }

        body {
            background-color: #0d1117; /* dark background */
            color: #e6edf3;            /* light text */
            font-family: Georgia, serif;
            padding: 20px;
        }

        h1 {
            font-size: 2rem;           /* rem = relative to root font size */
            margin-bottom: 8px;
            color: #3fb950;            /* green accent */
        }

        p {
            margin-bottom: 12px;
            line-height: 1.6;          /* spacing between lines */
        }
    </style>
</head>
<body>
    <h1>Text Monopoly</h1>
    <p>Welcome to the game! This page now has a dark theme.</p>
</body>
</html>
"""

filepath = os.path.join(project_folder, "01_styled_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"Styled page saved to: {filepath}")
print("Open it in your browser to see the dark theme.")
```

## Section 3 — CSS Selectors Explained

A **selector** tells the browser which elements to style. Here are the three most common types:

| Selector | Example | What it targets |
|---|---|---|
| Element | `h1 { ... }` | All `<h1>` tags |
| Class | `.subtitle { ... }` | Any element with `class="subtitle"` |
| ID | `#console { ... }` | The one element with `id="console"` |

Classes (`.`) are the most flexible because you can reuse them on many elements. IDs (`#`) must be unique on a page.

## Section 4 — Using CSS Variables for a Theme

CSS custom properties (also called variables) let you define colours once and reuse them everywhere. This makes it easy to change your whole colour scheme in one place. They always start with `--`.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Text Monopoly - CSS Variables</title>
    <style>
        /* Define our colour palette in one place */
        :root {
            --bg: #0d1117;
            --surface: #161b22;
            --border: #30363d;
            --text: #e6edf3;
            --dim: #8b949e;
            --green: #3fb950;
            --blue: #58a6ff;
            --red: #f85149;
        }

        * { margin: 0; padding: 0; box-sizing: border-box; }

        body {
            background: var(--bg);     /* use a variable with var() */
            color: var(--text);
            font-family: Georgia, serif;
            padding: 24px;
        }

        h1 { color: var(--green); margin-bottom: 6px; }

        .subtitle {
            color: var(--dim);         /* class selector */
            font-size: 0.9rem;
            margin-bottom: 20px;
        }

        .card {
            background: var(--surface);
            border: 1px solid var(--border);
            border-radius: 12px;       /* rounded corners */
            padding: 16px;
            margin-bottom: 12px;
        }

        .card h2 {
            font-size: 1.1rem;
            margin-bottom: 8px;
        }
    </style>
</head>
<body>
    <h1>Text Monopoly</h1>
    <p class="subtitle">Phase 1 — learning CSS variables and cards</p>

    <div class="card">
        <h2>Player 1</h2>
        <p>Cash: &pound;1500</p>
    </div>

    <div class="card">
        <h2>Player 2</h2>
        <p>Cash: &pound;1500</p>
    </div>
</body>
</html>
"""

filepath = os.path.join(project_folder, "01_styled_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Page updated with CSS variables and cards! Refresh your browser.")
```

## Section 5 — The Box Model

Every HTML element is a rectangular box. CSS controls four layers around the content:

1. **Content** — the text or image itself
2. **Padding** — space between the content and the border
3. **Border** — the visible edge
4. **Margin** — space outside the border, separating it from other elements

When you wrote `padding: 16px` and `margin-bottom: 12px` on the card, you were controlling these layers. The `box-sizing: border-box` rule we set on `*` makes padding count *inside* the width, which is much easier to work with.

## Section 6 — Styling Text and Buttons

Let us add a styled button. Buttons are essential for our game (rolling dice, buying properties). Notice how `:hover` creates an interactive effect when the user moves their mouse over the button.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Text Monopoly - Buttons</title>
    <style>
        :root {
            --bg: #0d1117;
            --surface: #161b22;
            --border: #30363d;
            --text: #e6edf3;
            --dim: #8b949e;
            --green: #3fb950;
        }

        * { margin: 0; padding: 0; box-sizing: border-box; }
        body {
            background: var(--bg);
            color: var(--text);
            font-family: Georgia, serif;
            padding: 24px;
        }
        h1 { color: var(--green); margin-bottom: 16px; }

        /* A styled button */
        .btn {
            display: inline-block;
            padding: 12px 24px;
            border: none;
            border-radius: 8px;
            font-size: 1rem;
            font-weight: bold;
            cursor: pointer;              /* hand icon on hover */
            transition: background 0.2s;  /* smooth colour change */
        }

        .btn-primary {
            background: var(--green);
            color: #fff;
        }

        /* :hover triggers when the mouse is over the element */
        .btn-primary:hover {
            background: #2ea043;
        }

        .btn:disabled {
            opacity: 0.4;
            cursor: not-allowed;
        }
    </style>
</head>
<body>
    <h1>Text Monopoly</h1>
    <button class="btn btn-primary">Roll Dice</button>
    <button class="btn btn-primary" disabled>Buy Property</button>
</body>
</html>
"""

filepath = os.path.join(project_folder, "01_styled_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Page updated with styled buttons! Refresh your browser.")
print("Hover over the Roll Dice button to see the colour change.")
```

## Section 7 — Layout with CSS Grid

CSS Grid is a powerful layout system. We will use it in our game to place the console on the left and the sidebar on the right. The key property is `display: grid` combined with `grid-template-columns`.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Text Monopoly - Grid Layout</title>
    <style>
        :root {
            --bg: #0d1117;
            --surface: #161b22;
            --border: #30363d;
            --text: #e6edf3;
            --dim: #8b949e;
            --green: #3fb950;
            --blue: #58a6ff;
        }

        * { margin: 0; padding: 0; box-sizing: border-box; }
        body {
            background: var(--bg);
            color: var(--text);
            font-family: Georgia, serif;
            padding: 24px;
        }
        h1 { color: var(--green); margin-bottom: 16px; }

        /* Two-column grid: main area takes remaining space, sidebar is 300px */
        .game-container {
            display: grid;
            grid-template-columns: 1fr 300px;
            gap: 20px;
            max-width: 1000px;
        }

        .console {
            background: var(--surface);
            border: 1px solid var(--border);
            border-radius: 12px;
            padding: 16px;
            min-height: 300px;
        }

        .sidebar {
            background: var(--surface);
            border: 1px solid var(--border);
            border-radius: 12px;
            padding: 16px;
        }
    </style>
</head>
<body>
    <h1>Text Monopoly</h1>

    <div class="game-container">
        <div class="console">
            <p>Game log will appear here...</p>
        </div>
        <div class="sidebar">
            <p>Player info and actions here...</p>
        </div>
    </div>
</body>
</html>
"""

filepath = os.path.join(project_folder, "01_styled_page.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Two-column grid layout saved! Refresh your browser.")
print("Resize the window to see how the columns behave.")
```
