# Workshop 02 — JavaScript Fundamentals

HTML gives your page structure, CSS makes it look good, and now **JavaScript** (JS) brings it to life. JavaScript is the programming language of the web. It runs right inside the browser and lets you respond to button clicks, update text on screen, and build game logic.

**What you will learn:**
- Variables, data types, and basic operations
- Functions and how to call them
- How JavaScript interacts with HTML elements
- Console logging for debugging

**Time:** roughly 15 to 20 minutes.

## Section 1 — Where Does JavaScript Go?

Just like CSS, JavaScript can go in three places:

1. **Inline** — inside an HTML attribute like `onclick="..."`. Quick but limited.
2. **Internal** — inside a `<script>` tag, usually just before `</body>`.
3. **External** — in a separate `.js` file. Best for larger projects.

We will use the **internal** method to keep everything in one file. The `<script>` tag goes at the bottom of the body so that all the HTML elements are loaded before our code tries to find them.

## Section 2 — Variables and Data Types

In JavaScript, you store values in **variables**. Use `let` for values that can change and `const` for values that should stay fixed. Let us write a page that demonstrates this.

```python
import os

project_folder = os.path.join(os.path.expanduser("~"), "text_monopoly")
os.makedirs(project_folder, exist_ok=True)

html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>JS Fundamentals</title>
    <style>
        body { background: #0d1117; color: #e6edf3; font-family: Georgia, serif; padding: 24px; }
        h1 { color: #3fb950; margin-bottom: 16px; }
        #output { background: #161b22; padding: 16px; border-radius: 8px; white-space: pre-line; }
    </style>
</head>
<body>
    <h1>JavaScript Variables</h1>
    <div id="output"></div>

    <script>
        // const = constant (cannot be reassigned)
        const gameName = "Text Monopoly";
        const startingCash = 1500;

        // let = can be changed later
        let playerPosition = 0;
        let inJail = false;

        // Display in the page
        const output = document.getElementById("output");
        output.textContent =
            "Game: " + gameName + "\n" +
            "Starting cash: £" + startingCash + "\n" +
            "Position: " + playerPosition + "\n" +
            "In jail: " + inJail;
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "02_js_fundamentals.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"File saved: {filepath}")
print("Open it in your browser. You should see the variable values displayed.")
```

## Section 3 — Arrays and Objects

Games need to store collections of data. An **array** is an ordered list. An **object** groups related values together using key-value pairs. In our game, the board is an array of objects.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Arrays and Objects</title>
    <style>
        body { background: #0d1117; color: #e6edf3; font-family: Georgia, serif; padding: 24px; }
        h1 { color: #3fb950; margin-bottom: 16px; }
        #output { background: #161b22; padding: 16px; border-radius: 8px; white-space: pre-line; }
    </style>
</head>
<body>
    <h1>Arrays and Objects</h1>
    <div id="output"></div>

    <script>
        // An array of property objects
        const properties = [
            { name: "Old Kent Road",  cost: 60,  rent: 10 },
            { name: "Whitechapel Rd", cost: 60,  rent: 10 },
            { name: "Mayfair",        cost: 400, rent: 100 }
        ];

        // Access array items with [index] (starts at 0)
        const first = properties[0];

        // Access object values with .key
        const output = document.getElementById("output");
        output.textContent =
            "First property: " + first.name + "\n" +
            "Cost: £" + first.cost + "\n" +
            "Rent: £" + first.rent + "\n\n" +
            "Total properties: " + properties.length;
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "02_js_fundamentals.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Updated with arrays and objects. Refresh your browser.")
```

## Section 4 — Functions

A **function** is a reusable block of code. You define it once and call it whenever you need it. In our game, actions like rolling dice and buying properties will each be their own function.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Functions</title>
    <style>
        body { background: #0d1117; color: #e6edf3; font-family: Georgia, serif; padding: 24px; }
        h1 { color: #3fb950; margin-bottom: 16px; }
        #output { background: #161b22; padding: 16px; border-radius: 8px; white-space: pre-line; }
        button {
            padding: 10px 20px; border: none; border-radius: 8px;
            background: #3fb950; color: #fff; font-size: 1rem;
            cursor: pointer; margin-bottom: 12px;
        }
        button:hover { background: #2ea043; }
    </style>
</head>
<body>
    <h1>Functions</h1>
    <button onclick="rollDice()">Roll Dice</button>
    <div id="output">Click the button to roll...</div>

    <script>
        // A function that simulates rolling two six-sided dice
        function rollDice() {
            // Math.random() gives 0 to 0.999...
            // Math.floor() rounds down to a whole number
            const die1 = Math.floor(Math.random() * 6) + 1;
            const die2 = Math.floor(Math.random() * 6) + 1;
            const total = die1 + die2;

            const output = document.getElementById("output");
            output.textContent =
                "Die 1: " + die1 + "\n" +
                "Die 2: " + die2 + "\n" +
                "Total: " + total;
        }
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "02_js_fundamentals.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Functions demo saved. Click the Roll Dice button multiple times!")
```

## Section 5 — Conditions (if / else)

Games are full of decisions. "If the player lands on a property, check if it is owned." JavaScript uses `if`, `else if`, and `else` to handle these branches.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Conditions</title>
    <style>
        body { background: #0d1117; color: #e6edf3; font-family: Georgia, serif; padding: 24px; }
        h1 { color: #3fb950; margin-bottom: 16px; }
        #output { background: #161b22; padding: 16px; border-radius: 8px; white-space: pre-line; }
        button {
            padding: 10px 20px; border: none; border-radius: 8px;
            background: #3fb950; color: #fff; font-size: 1rem;
            cursor: pointer; margin-bottom: 12px;
        }
    </style>
</head>
<body>
    <h1>Conditions</h1>
    <button onclick="checkSpace()">Land on Random Space</button>
    <div id="output">Click the button to land somewhere...</div>

    <script>
        const spaces = [
            { name: "GO",            type: "go" },
            { name: "Old Kent Road", type: "property", cost: 60 },
            { name: "Income Tax",    type: "tax", amount: 200 },
            { name: "GO TO JAIL",    type: "gotojail" },
            { name: "Free Parking",  type: "free" }
        ];

        function checkSpace() {
            // Pick a random space
            const index = Math.floor(Math.random() * spaces.length);
            const space = spaces[index];
            let message = "Landed on: " + space.name + "\n";

            if (space.type === "go") {
                message += "Collect £200!";
            } else if (space.type === "property") {
                message += "Available to buy for £" + space.cost;
            } else if (space.type === "tax") {
                message += "Pay £" + space.amount + " in tax!";
            } else if (space.type === "gotojail") {
                message += "Go directly to jail!";
            } else {
                message += "Nothing happens here. Relax!";
            }

            document.getElementById("output").textContent = message;
        }
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "02_js_fundamentals.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Conditions demo saved! Click the button several times.")
```

## Section 6 — Loops

Loops let you repeat an action. The `for` loop is the most common. We will use loops to iterate over the board spaces and player properties.

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Loops</title>
    <style>
        body { background: #0d1117; color: #e6edf3; font-family: Georgia, serif; padding: 24px; }
        h1 { color: #3fb950; margin-bottom: 16px; }
        #output { background: #161b22; padding: 16px; border-radius: 8px; white-space: pre-line; }
    </style>
</head>
<body>
    <h1>Loops</h1>
    <div id="output"></div>

    <script>
        const board = [
            { name: "GO",            type: "go" },
            { name: "Old Kent Road", type: "property" },
            { name: "Community Chest", type: "chest" },
            { name: "Whitechapel Rd", type: "property" },
            { name: "Income Tax",    type: "tax" }
        ];

        let text = "Board spaces:\n";

        // Loop through each space
        for (let i = 0; i < board.length; i++) {
            text += i + ": " + board[i].name +
                    " (" + board[i].type + ")\n";
        }

        // Count how many are properties
        let propertyCount = 0;
        for (let i = 0; i < board.length; i++) {
            if (board[i].type === "property") {
                propertyCount++;
            }
        }
        text += "\nTotal properties: " + propertyCount;

        document.getElementById("output").textContent = text;
    </script>
</body>
</html>
"""

filepath = os.path.join(project_folder, "02_js_fundamentals.html")
with open(filepath, "w", encoding="utf-8") as f:
    f.write(html_content)

print("Loops demo saved! Refresh your browser.")
```

## Section 7 — Interacting with the DOM

The **DOM** (Document Object Model) is how JavaScript sees and changes your HTML. You have already used `document.getElementById()` to find elements. Here are the key methods you will use throughout the game:

| Method | What it does |
|---|---|
| `document.getElementById("x")` | Finds the element with `id="x"` |
| `element.textContent = "..."` | Changes the text inside an element |
| `element.innerHTML = "..."` | Changes the HTML inside an element |
| `element.disabled = true` | Disables a button |
| `element.className = "..."` | Changes the CSS class(es) on an element |

These are the building blocks of every UI update in our game.
